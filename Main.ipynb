**توزیع نمایی :‌ زمان بین دو رخداد در یک فرایند تصادفی**

نکته‌ی خیلی مهم:
توزیع نمایی خودِ تعداد رخدادها را مدل نمی‌کند، بلکه زمان بین رخدادها را مدل می‌کند 

معمولاً پواسون → تعداد رخدادها در بازه زمانی 

معمولاً نمایی → زمان بین دو رخداد

توزیع نمایی تنها توزیع پیوسته ای هست که خاصیت بی حافظگی دارد

به صورت شهودی میشه گفت "گذشته، آینده را پیش‌بینی نمی‌کند" از نظر ریاضیاتی داریم که‌:

$$
P(X > s + t | X > s) = P(X > t)
$$

تفاوت PDF و CDF در توزیع نمایی
بعضی‌ها این دو را قاطی می‌کنند.

PDF:
$$
f(x)=λe^{−λx}
$$
 
این «چگالی» است، نه احتمال مستقیم روی یک نقطه.

CDF :
$$
F(x)=P(X≤x)=1−e^{−λx} 
$$
این احتمال واقعی تا یک مقدار است.

------------------------------------------------
نکته مهم : 

برای متغیر پیوسته:
$$
P(X=a)=0
$$
پس همیشه احتمال‌ها را روی بازه حساب می‌کنیم.

مثلاً:
$$
P(2<X<5)=F(5)−F(2)
$$

یک جمع‌بندی خیلی ساده
توزیع نمایی برای زمان انتظار تا رخداد بعدی است.

$$
f(x)=λe^{−λx} ,x≥0
$$
$$
F(x)=1−e^{−λx}
$$


$$
P(X>x)=e^{−λx}
$$



$$
E[X]= \frac{{1}}{{λ}}
​$$


 
$$
Var(X) = \frac{{1}}{{λ^{2}}}
​$$

$$
P(X>s+t∣X>s)=P(X>t)
