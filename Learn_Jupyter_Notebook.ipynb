**how to draw a chart :**

```python
import numpy as np
import matplotlib.pyplot as plt

throws = np.random.choice(["Head", "Tail"], size=100)

heads_count = np.sum(throws == "Head")
tails_count = np.sum(throws == "Tail")

plt.bar(["Head", "Tail"], [heads_count, tails_count])
plt.title("Coin Toss Simulation")
plt.show()
```

**How to write a mathematical formula :**

The probability of event A is:

$$
P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total outcomes}}
$$
